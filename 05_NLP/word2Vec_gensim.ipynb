```python
import gensim
from gensim.models import Word2Vec
import pandas as pd
```

```python
sentences=[
    ["i","love","machine","learning"],
    ["it","is","fun","to","learn","ml"],
    ["deep","learning","brings","thrill"],
    ["nlp","is","a","part","of","machine","learning"],
    ["word2vec","is","a","model"],
]
```

```python
model=Word2Vec(
    sentences,
    vector_size=50,
    window=3,
    min_count=1,
    workers=4
)
```

```python
print("Vocubulary:",list(model.wv.index_to_key))
```

50 numbers vector for 'learning' word

```python
print("\nVector for learning:\n",model.wv["learning"])
```

finding similar words

```python
print("words similar to learning:\n")
print(model.wv.most_similar("learning"))
```

find similarity between two words

```python
print("similarity between machine and learning")
print(model.wv.similarity("machine","learning"))
```

**Uisng a real world dataset on word2Vec**

```python
import nltk
from nltk.corpus import movie_reviews
from gensim.models import Word2Vec
nltk.download('movie_reviews')
```

```python
sentences=[]
for fileid in movie_reviews.fileids():
  words=movie_reviews.words(fileid)
  sentences.append([w.lower() for w in words])

print("Total Sentences:",len(sentences))
```

```python
model=Word2Vec(
    sentences,
    vector_size=100,
    window=5,
    min_count=2,
    workers=4
)
```

```python
print("Sample Vocabulary:",model.wv.index_to_key[:10])
```

```python
print("words similar to 'good':\n")
similar_words=model.wv.most_similar("good")
for word,score in similar_words:
  print(word,":",score)
#print(model.wv.most_similar("good"))
```

```python
print("Similarity between good and bad:")
print(model.wv.similarity("good","bad"))
```

**For better preprocessing using RE**

```python
import re
clean_sentences=[]
for sentence in sentences:
  cleaned=[re.sub(r'[^a-z]','',w) for w in sentence]
  cleaned=[w for w in cleaned if w!=""]
  clean_sentences.append(cleaned)
```

```python
model=Word2Vec(
    clean_sentences,
    vector_size=50,
    window=5,
    min_count=2,
    workers=4
)
```

```python
model.wv.most_similar("movie")
```

```python
model.wv.doesnt_match(['movie','film','banana'])
```
