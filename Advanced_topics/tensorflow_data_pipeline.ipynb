create tf dataset froma list

```python
daily_sales_numbers=[21,22,-108,31,-1,32,34,31]
tf_dataset=tf.data.Dataset.from_tensor_slices(daily_sales_numbers)
tf_dataset
```

iterate through tf dataset

```python
for sales in tf_dataset:
  print(sales.numpy())
```

iterate through elements as numpy elements

```python
for sales in tf_dataset.as_numpy_iterator():
  print(sales)
```

iterate through first n elements in tf dataset using take()=head()

```python
for sales in tf_dataset.take(3):
  print(sales.numpy())
```

filter sales numbers that are less than zero remove them

```python
tf_dataset=tf_dataset.filter(lambda x:x>0)
for sales in tf_dataset.as_numpy_iterator():
  print(sales)
```

these are in US dollar convert them into indian rupees

```python
tf_dataset=tf_dataset.map(lambda x:x*72)
for sales in tf_dataset.as_numpy_iterator():
  print(sales)
```

Shuffle them

```python
tf_dataset=tf_dataset.shuffle(2)
for sales in tf_dataset.as_numpy_iterator():
  print(sales)
```

Batching the data randomly

```python
for sales_batch in tf_dataset.batch(2):
  print(sales_batch.numpy())
```

Now performing all the above operations in one line--this is the
**tf data-pipeline**

```python
tf_dataset=tf.data.Dataset.from_tensor_slices(daily_sales_numbers)
tf_dataset=tf_dataset.filter(lambda x:x>0).map(lambda y:y*72).shuffle(2).batch(2)
for sales in tf_dataset.as_numpy_iterator():
  print(sales)
```

```python
import tensorflow as tf
import pathlib

data_dir = pathlib.Path("/root/.keras/datasets/flower_photos/flower_photos")

IMG_SIZE = 224
BATCH_SIZE = 32
```

```python
file_paths = tf.data.Dataset.list_files(str(data_dir / "*/*"), shuffle=True)

for file in file_paths.take(3):
    print(file.numpy())
```

```python
class_names = sorted([item.name for item in data_dir.glob('*')])
print("Classes:", class_names)
```

```python
def get_label(file_path):
    parts = tf.strings.split(file_path, '/')
    return parts[-2] == class_names
```

```python
def decode_img(img):
    img = tf.image.decode_jpeg(img, channels=3)
    img = tf.image.resize(img, [IMG_SIZE, IMG_SIZE])
    return img / 255.0
```

```python
def process_path(file_path):
    label = get_label(file_path)
    img = tf.io.read_file(file_path)
    img = decode_img(img)
    return img, label
```

```python
dataset = file_paths.map(process_path, num_parallel_calls=tf.data.AUTOTUNE)
```

```python
dataset = dataset.shuffle(1000)
dataset = dataset.batch(BATCH_SIZE)
dataset = dataset.prefetch(buffer_size=tf.data.AUTOTUNE)
```

```python
dataset_size = len(list(data_dir.glob('*/*')))

train_size = int(0.8 * dataset_size)

train_ds = dataset.take(train_size // BATCH_SIZE)
val_ds = dataset.skip(train_size // BATCH_SIZE)
```

```python
from tensorflow.keras import layers, models

model = models.Sequential([
    layers.Conv2D(32, 3, activation='relu', input_shape=(224,224,3)),
    layers.MaxPooling2D(),

    layers.Conv2D(64, 3, activation='relu'),
    layers.MaxPooling2D(),

    layers.Conv2D(128, 3, activation='relu'),
    layers.MaxPooling2D(),

    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(len(class_names), activation='softmax')
])
```

```python
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=5
)
```
